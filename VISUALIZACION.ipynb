```python
import pandas as pd
import matplotlib.pyplot as plt

# (por si acaso, vuelve a cargar)
df = pd.read_csv("SampleSuperstore.csv", encoding="latin-1")

# crear variable otra vez
sales_cat = df.groupby('Category')['Sales'].sum().sort_values(ascending=False)

# gráfico
sales_cat.plot(kind='bar')
plt.title("Ventas por categoría")
plt.show()
```

```python
# análisis completo en orden
df = pd.read_csv("SampleSuperstore.csv", encoding="latin-1")

sales_cat = df.groupby('Category')['Sales'].sum()
profit_cat = df.groupby('Category')['Profit'].sum()
sub_profit = df.groupby('Sub-Category')['Profit'].sum()
discount_profit = df.groupby('Discount')['Profit'].mean()
```

```python
sub_profit.plot(kind='bar')
plt.title("Ganancia por subcategoría")
plt.show()
```

```python
discount_profit.plot()
plt.title("Descuento vs Ganancia")
plt.show()
```
