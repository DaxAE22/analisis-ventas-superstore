IMPORTAMOS EL DATASET DEL PROYECTO


```python
import pandas as pd


df = pd.read_csv("SampleSuperstore.csv")


print(df.head())
```

MOSTRAMOS LAS COLUMNAS EXISTENTE PARA PODER HACER UN MEJOR ANALISIS



```python
print(df.columns)
```

```python
print(df.head())
print(df.info())
print(df.describe())
```

REVISAMOS SI EXISTEN VALORES NULOS

```python
print(df.isnull().sum())
```

REVISAMOS LOS DUPLICADOS

```python
print(df.duplicated().sum())
```

LIMPIAMOS LOS DUPLICADOS

```python
df = df.drop_duplicates()
```

Ventas y ganancias por categoría

```python
sales_cat = df.groupby('Category')['Sales'].sum().sort_values(ascending=False)
profit_cat = df.groupby('Category')['Profit'].sum().sort_values(ascending=False)

print("Ventas por categoría:\n", sales_cat)
print("\nGanancia por categoría:\n", profit_cat)
```

Subcategorías problemáticas

```python
sub_profit = df.groupby('Sub-Category')['Profit'].sum().sort_values()
print(sub_profit)
```

Impacto del descuento

```python
discount_profit = df.groupby('Discount')['Profit'].mean()
print(discount_profit)
```

Regiones más rentables

```python
region_profit = df.groupby('Region')['Profit'].sum().sort_values(ascending=False)
print(region_profit)
```

Top productos

```python
top_products = df.groupby('Sub-Category')['Sales'].sum().sort_values(ascending=False).head(10)
print(top_products)
```
